```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv(r"C:\Users\KARAN\Downloads\DSML_PRACTICAL\DSML_PRACTICAL\Datasets\IRIS.csv")
```

```python
numerical_features = ["sepal_length", "sepal_width", "petal_length", "petal_width"]
```

```python
# for feature in numerical_features:
#     sns.boxplot(x=data[feature], color='red')
#     plt.title(f"Box plot of {feature}")
#     plt.xlabel(feature)
#     plt.show()



for feature in numerical_features:
    Q1 = data[feature].quantile(0.25)
    Q3 = data[feature].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = data[(data[feature] < lower_bound) | (data[feature] > upper_bound)]
    print(f"Outliers in {feature}:\n", outliers[feature])
    
    # Plotting the boxplot
    sns.boxplot(x=data[feature], color='red')
    plt.title(f"Box plot of {feature}")
    plt.xlabel(feature)
    plt.show()
```
