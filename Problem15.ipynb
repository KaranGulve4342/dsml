```python
# 15. Use the dataset 'titanic'. The dataset contains 891 rows and contains
# information about the passengers who boarded the unfortunate Titanic
# ship. Use the Seaborn library to see if we can find any patterns in the data.

import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv(r"C:\Users\KARAN\Downloads\DSML_PRACTICAL\DSML_PRACTICAL\Datasets\Titanic.csv")
```

```python
data.head()
```

```python
## Survival rate based on Gender

plt.figure(figsize=(8,6))
sns.countplot(x='Sex', hue='Survived', data=data)
plt.title("Survival rate based on gender")
# plt.show()
```

```python
## Survival Rate based on Passenger Class
plt.figure(figsize=(8,6))
sns.countplot(x='Pclass', hue='Survived', data=data)
plt.title('Survival Rate based on passenger class')
plt.show()
```

```python
# Age Distribution of Survivors and Non-Survivors

plt.figure(figsize=(8, 6))
sns.histplot(data=data, x='Age', hue='Survived', multiple='stack', kde=True)
plt.title("Age Distribution of Survivors and Non-Survivors")
plt.show()
```

```python
## Survival Rate based on Embarked Port

plt.figure(figsize=(8, 6))
sns.countplot(x='Embarked', hue='Survived', data=data)
plt.title('Survival Rate Based on Embarked Port')
plt.show()
```

```python
## Pairplot to see Relationship between features
sns.pairplot(data.dropna(), hue='Survived', diag_kind='kde')
plt.show()
```
