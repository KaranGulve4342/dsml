```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv(r"C:\Users\KARAN\Downloads\DSML_PRACTICAL\DSML_PRACTICAL\Datasets\IRIS.csv")
```

```python
data.head()
```

```python
print("Features and their types: ")
for column in data.columns[:-1]: ## Exclude species columns
    print(f"{column}: {data[column].dtype}")
```

```python
## Create Histogram for each features

plt.figure(figsize=(10,8))
features=['sepal_length', 'sepal_width', 'petal_length', 'petal_width']

for i, feature in enumerate(features):
    plt.subplot(2, 2, i+1)
    sns.histplot(data[feature], kde=True, bins=20, color='green')
    plt.title(f"Distribution of {feature}")
    plt.xlabel(feature)
    plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```

```python
## Create a Box plot for each feature in the dataset

features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']

for feature in features:
    sns.boxplot(x='species', y=feature, data=data)
    plt.title(f"Box plot of {feature} vs Species")
    plt.show()
```

```python
grouped = data.groupby('species')
```

```python
for species, data in grouped:
    print(f"Statistics for {species}:")
    print(data.describe())
    print("\n")
```
