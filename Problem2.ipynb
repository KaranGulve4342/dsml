```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r'C:\Users\KARAN\Downloads\DSML_PRACTICAL\DSML_PRACTICAL\Datasets\Telecom Churn.csv')
```

```python
print("First few rows of the dataset:")
print(df.head())
```

```python
## Summary statistics for each feature
for column in df.select_dtypes(include=[np.number]).columns:
    print(f"\nStatistics for {column}:")

    # Minimum value
    print(f"Minimum: {df[column].min()}")

    # Maximum value
    print(f"Maximum: {df[column].max()}")

    # Mean value
    print(f"Mean: {df[column].mean()}")

    # Range of values
    data_range = df[column].max() - df[column].min()
    print(f"Range: {data_range}")

    # Standard Deviation
    print(f"Standard Deviation: {df[column].std()}")

    # Variance
    print(f"Variance: {df[column].var()}")

    # Percentiles
    percentiles = df[column].quantile([0.25, 0.50, 0.75]).to_dict()
    print(f"25th Percentile: {percentiles[0.25]}")
    print(f"50th Percentile (Median): {percentiles[0.50]}")
    print(f"75th Percentile: {percentiles[0.75]}")
```
