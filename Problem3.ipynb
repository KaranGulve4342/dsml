```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("Datasets/House Data.csv")
```

```python
df.head()
```

```python
print(df.info())
```

```python
df.drop(['Unnamed: 0'], axis=1, inplace=True)
```

```python
for column in df.select_dtypes(include=[np.number]).columns:
    print(f"\nFeature distribution for {column}")

    print(f"\nStandard Deviation: {df[column].std()}")

    print(f"\nVariance: {df[column].var()}")

    percentiles = df[column].quantile([0.25, 0.50, 0.75]).to_dict()

    print(f"25th percentile: {percentiles[0.25]}")
    print(f"50th percentile: {percentiles[0.50]}")
    print(f"75th percentile: {percentiles[0.75]}")
```

```python
## Create Histograms for each Numerical Feature using seaborn

for column in df.select_dtypes(include=[np.number]).columns:
    plt.figure(figsize=(8, 6))

    sns.histplot(df[column])

    plt.title(f"Histogram of {column}", fontsize=16)

    plt.xlabel(column, fontsize=14)

    plt.ylabel("Frequency", fontsize=14)

    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()
```
